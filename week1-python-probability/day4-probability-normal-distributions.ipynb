```python
import numpy as np
import matplotlib.pyplot as plt

samples = np.random.normal(loc=0, scale=1, size=10000)

# Create histogram
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=50, density=True, alpha=0.7, edgecolor='black', color='skyblue')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Standard Normal Distribution N(0, 1) - 10,000 Samples')
plt.grid(True, alpha=0.3)
plt.show()

# Compute statistics
print(f"Sample mean: {np.mean(samples):.4f} ")
print(f"Sample std: {np.std(samples):.4f} ")
print(f"Sample min: {np.min(samples):.4f}")
print(f"Sample max: {np.max(samples):.4f}")
```

```python

mu_values = [-2, 0, 2, 4]
sigma = 1  

plt.figure(figsize=(12, 6))

for mu in mu_values:
    samples = np.random.normal(loc=mu, scale=sigma, size=10000)
    plt.hist(samples, bins=50, density=True, alpha=0.5, 
             label=f'mean={mu}, std={sigma}')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution: Effect of Mean')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
