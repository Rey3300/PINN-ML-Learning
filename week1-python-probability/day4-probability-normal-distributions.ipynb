```python
import numpy as np
import matplotlib.pyplot as plt

samples = np.random.normal(loc=0, scale=1, size=10000)

# Create histogram
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=50, density=True, alpha=0.7, edgecolor='black', color='skyblue')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Standard Normal Distribution N(0, 1) - 10,000 Samples')
plt.grid(True, alpha=0.3)
plt.show()

# Compute statistics
print(f"Sample mean: {np.mean(samples):.4f} ")
print(f"Sample std: {np.std(samples):.4f} ")
print(f"Sample min: {np.min(samples):.4f}")
print(f"Sample max: {np.max(samples):.4f}")
```

```python

mu_values = [-2, 0, 2, 4]
sigma = 1  

plt.figure(figsize=(12, 6))

for mu in mu_values:
    samples = np.random.normal(loc=mu, scale=sigma, size=10000)
    plt.hist(samples, bins=50, density=True, alpha=0.5, 
             label=f'mean={mu}, std={sigma}')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution: Effect of Mean')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
mean = 0  
std_values = [0.5, 1, 2, 3]

plt.figure(figsize=(12, 6))

for std in std_values:
    samples = np.random.normal(loc=mean, scale=std, size=10000)
    plt.hist(samples, bins=50, density=True, alpha=0.5, 
             label=f'mean={mean}, std={std}')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution: Effect of Standard Deviation')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(-10, 10)
plt.show()

# What did you observe?
# - Larger   wider distribution (more spread)
# - Smaller   narrower distribution (more concentrated)
# - Center stays at  = 0
# - Peak gets lower as  increases (area under curve always = 1)
```

```python
# Comparing two students' scores on different tests using z-scores

score_A = 85
mean_A = 70
std_A = 10

score_B = 87
mean_B = 80
std_B = 5

z_score_A = (score_A - mean_A) / std_A
z_score_B = (score_B - mean_B) / std_B

print("Student 1 (Test A):")
print(f"  Raw score: {score_A}")
print(f"  Z-score: {z_score_A:.2f}")
print(f"  This is {z_score_A:.2f} standard deviations above the mean")
print()
print("Student 2 (Test B):")
print(f"  Raw score: {score_B}")
print(f"  Z-score: {z_score_B:.2f}")
print(f"  This is {z_score_B:.2f} standard deviations above the mean")
print()

if z_score_A > z_score_B:
    print("Student 1 performed better relative to their class")
else:
    print("Student 2 performed better relative to their class")

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Test A
test_A_scores = np.random.normal(mean_A, std_A, 10000)
axes[0].hist(test_A_scores, bins=50, density=True, alpha=0.7, edgecolor='black')
axes[0].axvline(mean_A, color='black', linestyle='--', linewidth=2, label='Mean')
axes[0].axvline(score_A, color='red', linestyle='-', linewidth=3, label=f'Student 1: {score_A}')
axes[0].set_xlabel('Test A Score')
axes[0].set_ylabel('Density')
axes[0].set_title(f'Test A Distribution (mean={mean_A}, std={std_A})')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Test B
test_B_scores = np.random.normal(mean_B, std_B, 10000)
axes[1].hist(test_B_scores, bins=50, density=True, alpha=0.7, edgecolor='black')
axes[1].axvline(mean_B, color='black', linestyle='--', linewidth=2, label='Mean')
axes[1].axvline(score_B, color='red', linestyle='-', linewidth=3, label=f'Student 2: {score_B}')
axes[1].set_xlabel('Test B Score')
axes[1].set_ylabel('Density')
axes[1].set_title(f'Test B Distribution (mean={mean_B}, std={std_B})')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python

# Question: For N(0,1), what is P(X > 1.5)?

samples = np.random.normal(0, 1, 100000)
prob_greater_than_1_5 = np.mean(samples > 1.5)

print(f"P(X > 1.5)  {prob_greater_than_1_5:.4f}")
print(f"Theoretical value: ~0.0668")

# Question: What is P(X < -1)?
prob_less_than_minus_1 = np.mean(samples < -1)
print(f"\nP(X < -1)  {prob_less_than_minus_1:.4f}")
print(f"Theoretical value: ~0.1587")

# Question: What is P(-0.5 < X < 0.5)?
prob_between = np.mean((samples > -0.5) & (samples < 0.5))
print(f"\nP(-0.5 < X < 0.5)  {prob_between:.4f}")
print(f"Theoretical value: ~0.3829")

# Visualize one of these
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=100, density=True, alpha=0.7, edgecolor='black')

# Shade region where X > 1.5
x_shade = samples[samples > 1.5]
plt.hist(x_shade, bins=50, density=True, alpha=0.8, color='red', 
         edgecolor='black', label=f'P(X > 1.5)  {prob_greater_than_1_5:.4f}')

plt.axvline(1.5, color='red', linestyle='--', linewidth=2)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Computing Probabilities from Normal Distribution')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
##If X ~ N(, ) and Y ~ N(, ) are independent,
##            then X + Y ~ N( + ,  + )

# Generate two independent normal variables
X = np.random.normal(5, 2, 10000)   # N(5, 4)
Y = np.random.normal(3, 1, 10000)   # N(3, 1)

# Sum them
Z = X + Y

# Check properties
print("X ~ N(5, 4):")
print(f"  Mean: {np.mean(X):.3f} (expected: 5)")
print(f"  Std: {np.std(X):.3f} (expected: 2)")
print()
print("Y ~ N(3, 1):")
print(f"  Mean: {np.mean(Y):.3f} (expected: 3)")
print(f"  Std: {np.std(Y):.3f} (expected: 1)")
print()
print("Z = X + Y should be ~ N(8, 5):")
print(f"  Mean: {np.mean(Z):.3f} (expected: 8)")
print(f"  Std: {np.std(Z):.3f} (expected: {np.sqrt(5):.3f})")
print(f"  Variance: {np.var(Z):.3f} (expected: 5)")

# Visualize all three
plt.figure(figsize=(14, 5))

plt.subplot(1, 3, 1)
plt.hist(X, bins=50, density=True, alpha=0.7, edgecolor='black')
plt.title('X ~ N(5, 4)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
plt.hist(Y, bins=50, density=True, alpha=0.7, edgecolor='black', color='orange')
plt.title('Y ~ N(3, 1)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 3)
plt.hist(Z, bins=50, density=True, alpha=0.7, edgecolor='black', color='green')
plt.title('Z = X + Y ~ N(8, 5)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Key insight: Sum of independent normals is also normal!
```

```python
# If X ~ N(, ), then aX + b ~ N(a + b, a)

X = np.random.normal(10, 3, 10000)  # N(10, 9)

a = 2
b = 5

Y = a * X + b  # Should be N(2*10 + 5, 4*9) = N(25, 36)

print(f"X ~ N(10, 9):")
print(f"  Mean: {np.mean(X):.3f}")
print(f"  Std: {np.std(X):.3f}")
print()
print(f"Y = 2X + 5 should be ~ N(25, 36):")
print(f"  Mean: {np.mean(Y):.3f} (expected: 25)")
print(f"  Std: {np.std(Y):.3f} (expected: 6)")
print(f"  Variance: {np.var(Y):.3f} (expected: 36)")

# Visualize
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.hist(X, bins=50, density=True, alpha=0.7, edgecolor='black')
ax1.axvline(np.mean(X), color='red', linestyle='--', linewidth=2, label=f'Mean: {np.mean(X):.1f}')
ax1.set_title('X ~ N(10, 9)')
ax1.set_xlabel('Value')
ax1.set_ylabel('Density')
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2.hist(Y, bins=50, density=True, alpha=0.7, edgecolor='black', color='orange')
ax2.axvline(np.mean(Y), color='red', linestyle='--', linewidth=2, label=f'Mean: {np.mean(Y):.1f}')
ax2.set_title('Y = 2X + 5 ~ N(25, 36)')
ax2.set_xlabel('Value')
ax2.set_ylabel('Density')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Key insight: Linear transformations of normal are still normal!
```
