```python
plt.plot(ages_x1, dev_y)
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salry (USD)")
plt.show()
```

```python
plt.plot(ages_x1, py_dev_y, 'k--')
plt.plot(ages_x1, dev_y, 'b')
plt.title("Median Python Developer Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salry (USD)")

plt.legend(["Python Devs", "All Devs"])
plt.show()
```

```python
plt.plot(ages_x1, py_dev_y, label = "Python devs", color="k", linestyle= "--", marker='.')
plt.plot(ages_x1, dev_y, label = "All devs", color="b", linestyle= "-", marker='o')
plt.title("Median Python Developer Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salry (USD)")


plt.show()
```

```python
plt.plot(ages_x1, py_dev_y, label = "Python devs", color="#444444", linestyle= "--", marker='.', linewidth = 3)
plt.plot(ages_x1, js_dev_y, label = "JavaScript devs", color="#adad3b", linestyle= "-", marker='v', linewidth = 3)
plt.plot(ages_x1, dev_y, label = "All devs", color="#5a7d9a", linestyle= "-", marker='o')
plt.title("Median Python Developer Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salry (USD)")

plt.legend()

plt.grid(True)
plt.tight_layout()

#plt.savefig('plot.png') # save the plot as a file
plt.show()
```
