```python
import numpy as np
import matplotlib.pyplot as plt

flips = np.random.choice(['H', 'T'], size = 10000)
prob_heads = np.mean(flips == 'H')

plt.hist((flips == 'H').astype(int), bins = np.arange(3)-0.5, density = True)
plt.xticks([0, 1], ['Tails', 'Heads'])

plt.ylabel('Frequency')
plt.title('10,000 Coin Flips')
plt.show()
```

```python
roll = np.random.randint(1, 7, size = 10000)
prob1 = np.mean(roll == 1)
prob2 = np.mean(roll == 2)
prob3 = np.mean(roll == 3)
prob4 = np.mean(roll == 4)
prob5 = np.mean(roll == 5)
prob6 = np.mean(roll == 6)
plt.hist(roll, bins = np.arange(1, 8)-0.5 , density = True)
plt.xlabel('Die Number')
plt.ylabel('Frequency')
plt.title('10,000 Die Rolls')
plt.show()
```

```python
#Subplots
fig, axes = plt.subplots(2,3, figsize = (12, 8))
axes[0,0].plot(prob1, 'ro')
axes[0,0].set_title('Probability of Rolling a 1')
axes[0,1].plot(prob2, 'go')
axes[0,1].set_title('Probability of Rolling a 2')
axes[0,2].plot(prob3, 'bo')
axes[0,2].set_title('Probability of Rolling a 3')
axes[1,0].plot(prob4, 'co')
axes[1,0].set_title('Probability of Rolling a 4')
axes[1,1].plot(prob5, 'mo')
axes[1,1].set_title('Probability of Rolling a 5')
axes[1,2].plot(prob6, 'yo')
axes[1,2].set_title('Probability of Rolling a 6')
plt.show()
```

```python
sample_sizes = [10, 50, 100, 500, 1000, 5000, 10000]
proportions = []

for n in sample_sizes:
    flips = np.random.choice(['H', 'T'], size=n)
    prop = np.mean(flips == 'H')
    proportions.append(prop)

plt.semilogx(sample_sizes, proportions, 'o-', markersize=10)
plt.axhline(0.5, color='red', linestyle='--', label='True probability')
plt.xlabel('Number of flips')
plt.ylabel('Proportion of heads')
plt.title('Convergence to True Probability')
plt.legend()
plt.grid(True)
plt.show()
```
